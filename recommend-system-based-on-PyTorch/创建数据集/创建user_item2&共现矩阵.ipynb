```python
import os
os.chdir("/content/drive/MyDrive/Colab Notebooks/推荐系统/推荐系统算法复现/ml-100k")
import numpy as np
import pandas as pd
```

```python
def loadData():
    f = open('u.data')
    data = []
    for i in range(100000):
        h = f.readline().split('\t')
        h = list(map(int, h))
        data.append(h[0:3])
    f.close()
    return data
```

```python
df = pd.DataFrame(loadData())
df.columns = ['user_id','item_id','rating']
df_sorted = df.sort_values(by='user_id', ascending=True)
# 将评分大于等于4的评分记为1，其他记为0
df_sorted.rating = df_sorted.rating.apply(lambda x:1 if x >=4 else 0)
df_sorted
```

```python
df_sorted.to_csv("user_item2.csv")
```

```python
# 生成用户和物体的共现矩阵
from scipy.sparse import coo_matrix

# 为用户和物品创建唯一的数字索引
user_ids = df_sorted['user_id'].astype("category").cat.codes
item_ids = df_sorted['item_id'].astype("category").cat.codes

# 创建共现矩阵
cooccurrence_matrix = coo_matrix(
    (df_sorted['rating'], (user_ids, item_ids)),
    shape=(df_sorted.user_id.nunique(), df_sorted.item_id.nunique())
)

co_mat = cooccurrence_matrix.toarray()

print(co_mat.shape) # (用户数量，物体数量)
print(co_mat)
```

```python
pd.DataFrame(co_mat).to_csv('co_mat2.csv')
```
