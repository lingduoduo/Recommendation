```python
# 加载rating数据
ratings = pd.read_csv('./u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
```

```python
ratings.rating = ratings.rating.apply(lambda x: 1 if x >= 4 else 0)
ratings
```

```python
grouped_user = ratings.groupby('user_id').agg(list)
grouped_user
```

```python
aa = grouped_user.iloc[0,:]
df = pd.DataFrame()
df['rating'] = aa.rating
df['item_id'] = aa.item_id
df['timestamp'] = aa.timestamp
table = df.sort_values(by = 'timestamp', ascending= True)
item_id_last1 = table.item_id.shift(1); table['item_id_last1'] = item_id_last1
item_id_last2 = table.item_id.shift(2); table['item_id_last2'] = item_id_last2
item_id_last3 = table.item_id.shift(3); table['item_id_last3'] = item_id_last3
item_id_last4 = table.item_id.shift(4); table['item_id_last4'] = item_id_last4
item_id_last5 = table.item_id.shift(5); table['item_id_last5'] = item_id_last5

rating_last1 = table.rating.shift(1); table['rating_last1'] = rating_last1
rating_last2 = table.rating.shift(2); table['rating_last2'] = rating_last2
rating_last3 = table.rating.shift(3); table['rating_last3'] = rating_last3
rating_last4 = table.rating.shift(4); table['rating_last4'] = rating_last4
rating_last5 = table.rating.shift(5); table['rating_last5'] = rating_last5
table.insert(1, 'user_id', 1)
table = table.dropna()
table.head()
```

```python
for i in range(1, grouped_user.shape[0]):
  bb = grouped_user.iloc[i,:]
  df = pd.DataFrame()
  df['rating'] = bb.rating
  df['item_id'] = bb.item_id
  df['timestamp'] = bb.timestamp
  table2 = df.sort_values(by = 'timestamp', ascending= True)
  item_id_last1 = table2.item_id.shift(1); table2['item_id_last1'] = item_id_last1
  item_id_last2 = table2.item_id.shift(2); table2['item_id_last2'] = item_id_last2
  item_id_last3 = table2.item_id.shift(3); table2['item_id_last3'] = item_id_last3
  item_id_last4 = table2.item_id.shift(4); table2['item_id_last4'] = item_id_last4
  item_id_last5 = table2.item_id.shift(5); table2['item_id_last5'] = item_id_last5

  rating_last1 = table2.rating.shift(1); table2['rating_last1'] = rating_last1
  rating_last2 = table2.rating.shift(2); table2['rating_last2'] = rating_last2
  rating_last3 = table2.rating.shift(3); table2['rating_last3'] = rating_last3
  rating_last4 = table2.rating.shift(4); table2['rating_last4'] = rating_last4
  rating_last5 = table2.rating.shift(5); table2['rating_last5'] = rating_last5

  table2.insert(1, 'user_id', i+1)
  table2 = table2.dropna()
  table = pd.concat([table, table2], axis = 0)
```
